```python
total = len(df)

success = (df["booking_status"] == "Success").sum()
cancel_driver = (df["booking_status"] == "Canceled by Driver").sum()
cancel_customer = (df["booking_status"] == "Canceled by Customer").sum()
driver_not_found = (df["booking_status"] == "Driver Not Found").sum()

print("Total Bookings:", total)
print("Success:", success)
print("Canceled by Driver:", cancel_driver)
print("Canceled by Customer:", cancel_customer)
print("Driver Not Found:", driver_not_found)

print("\nSuccess Rate %:", round(success/total*100, 2))
print("Cancellation Rate %:", round((cancel_driver+cancel_customer)/total*100, 2))
print("Driver Not Found %:", round(driver_not_found/total*100, 2))
```

```python
vehicle_summary = df.groupby("vehicle_type").agg(
    total_bookings=("booking_id", "count"),
    successful_rides=("booking_status", lambda x: (x == "Success").sum()),
    cancelled_by_driver=("booking_status", lambda x: (x == "Canceled by Driver").sum()),
    cancelled_by_customer=("booking_status", lambda x: (x == "Canceled by Customer").sum()),
    driver_not_found=("booking_status", lambda x: (x == "Driver Not Found").sum()),
   
    avg_fare=("booking_value", "mean"),
    avg_distance=("ride_distance", "mean"),
    unique_customers=("customer_id", "nunique"),
    avg_driver_rating=("driver_ratings", "mean")
).reset_index()

vehicle_summary["success_rate_%"] = (vehicle_summary["successful_rides"] / vehicle_summary["total_bookings"] * 100).round(2)

vehicle_summary.sort_values(by="successful_rides", ascending=False)
```

```python
success_df = df[df["booking_status"] == "Success"].copy()
success_df["revenue"] = success_df["booking_value"]

pickup_revenue = success_df.groupby("pickup_location")["revenue"].sum().sort_values(ascending=False)

top_pickup_location = pickup_revenue.index[0]

top_location_total_revenue = pickup_revenue.iloc[0]

top_location_avg_revenue = success_df.loc[
    success_df["pickup_location"] == top_pickup_location, "revenue"
].mean()

print("Top Pickup Location:", top_pickup_location)
print("Total Revenue:", top_location_total_revenue)
print("Average Revenue:", round(top_location_avg_revenue, 2))
```
